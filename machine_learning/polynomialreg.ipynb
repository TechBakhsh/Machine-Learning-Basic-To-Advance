```python
import numpy as np
import matplotlib.pyplot as plt

# Your data
x = [89, 43, 36, 95, 10, 66, 34, 38, 20, 26, 29, 48, 64, 6, 5]
y = [21, 46, 3, 35, 67, 95, 53, 72, 58, 10, 26, 34, 90, 33, 38]
model=x,y

# Fit a high-degree polynomial model (degree = number of data points - 1)
degree = len(x) - 1
model = np.poly1d(np.polyfit(x, y, degree))

# Generate a range of x values for prediction
line = np.linspace(2, 95, 100)

# Calculate the predicted y values
predicted_y = model(line)

# Plot the data and the fitted curve
plt.scatter(x, y)
plt.plot(line, predicted_y)
plt.title(f'{degree}-degree Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

# Visualization

```python

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
x = np.array([89, 43, 36, 95, 10, 66, 34, 38, 20, 26, 29, 48, 64, 6, 5])
y = np.array([21, 46, 3, 35, 67, 95, 53, 72, 58, 10, 26, 34, 90, 33, 38])

def viz_linear():
    # Create a linear regression model
    model = LinearRegression()
    
    # Reshape x to have a 2D shape, as scikit-learn expects a 2D array
    x_reshaped = x.reshape(-1, 1)
    
    # Fit the model to the data
    model.fit(x_reshaped, y)

    # Make predictions using the reshaped x
    y_pred = model.predict(x_reshaped)
    
    # Visualize the data and the regression line
    plt.scatter(x, y)
    plt.plot(x, y_pred, color="green")
    plt.xlabel("train")
    plt.ylabel("salary")
    plt.title("Linear Regression Model")
    plt.show()

viz_linear()
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
x = np.array([89, 43, 36, 95, 10, 66, 34, 38, 20, 26, 29, 48, 64, 6, 5])
y = np.array([21, 46, 3, 35, 67, 95, 53, 72, 58, 10, 26, 34, 90, 33, 38])

def viz_linear():
    # Create a linear regression model
    model = LinearRegression()
    
    # Reshape x to have a 2D shape, as scikit-learn expects a 2D array
    x_reshaped = x.reshape(-1, 1)
    
    # Fit the model to the data
    model.fit(x_reshaped, y)

    # Make predictions using the reshaped x
    y_pred = model.predict(x_reshaped)
    
    # Visualize the data and the regression line
    plt.scatter(x, y, label='Data')
    plt.plot(x, y_pred, color="green", label='Regression Line')
    plt.xlabel("train")
    plt.ylabel("salary")
    plt.title("Linear Regression Model")
    plt.legend()
    plt.show()

viz_linear()
```
